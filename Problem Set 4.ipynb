Question 2 (3 points):The final radius of a protostar ùëÖ_2 is a function of its mass and initial radius ùëÖ_1: 

$$GM(1/R_2 - 1/R_1) \approx 1/m_H (\epsilon_D/2 + \epsilon_I)$$

where, ùúÄ_ùê∑ = 4.5 eV and ùúÄ_ùêº = 13.6 eV

a. From the expression of the Jeans mass and Jeans radius, derive an expression for the Jeans density.

b. For a Jeans density of 10^‚àí16kg m^‚àí3 plot the initial radius of protostars between 0.2 and 2 solar masses

c. On a different graph, plot the final radius of the same protostars.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
G = 6.67430e-11  # m^3 kg^-1 s^-2
m_H = 1.6735575e-27  # kg
epsilon_D = 4.5 * 1.60218e-19  # convert eV to Joules
epsilon_I = 13.6 * 1.60218e-19  # convert eV to Joules
rho_jeans = 1e-16  # kg/m^3

# Mass range in kg
M_sun = 1.989e30  # kg
M = np.linspace(0.2, 2, 500) * M_sun

# Initial radius R1 assuming constant Jeans density
R1 = (3 * M / (4 * np.pi * rho_jeans)) ** (1 / 3)

# Final radius R2 using energy balance
delta_E = (epsilon_D / 2 + epsilon_I) / m_H
inv_R2 = 1 / R1 + delta_E / (G * M)
R2 = 1 / inv_R2

# Convert to solar radii for plotting
R_sun = 6.957e8  # m
R1_solar = R1 / R_sun
R2_solar = R2 / R_sun
M_solar = M / M_sun

# Plot R1 vs M
plt.figure(figsize=(8, 6))
plt.plot(M_solar, R1_solar, label='Initial Radius $R_1$', color='mediumpurple')
plt.xlabel('Mass [$M_\\odot$]')
plt.ylabel('Initial Radius [$R_\\odot$]')
plt.xlim(0.2,2)
plt.title('Initial Radius vs Mass')
plt.legend()
plt.tight_layout()
plt.show()

# Plot R2 vs M
plt.figure(figsize=(8, 6))
plt.plot(M_solar, R2_solar, label='Final Radius $R_2$', color='mediumpurple')
plt.xlabel('Mass [$M_\\odot$]')
plt.ylabel('Final Radius [$R_\\odot$]')
plt.title('Final Radius vs Mass')
plt.xlim(0.2,2)
plt.legend()
plt.tight_layout()
plt.show()
```
